# first work for jumia website

```python
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

page = 1
t = {"num_of_people_rated":[], 
     'Rating':[], 
     'discount': [], 
     'new_price':[], 
     'old_price':[], 
     'all_items_name':[]}

while page <= 50:
    url = "https://www.jumia.com.ng/catalog/?q=shoes&page={}#catalog-listing".format(page)
    html = requests.get(url)
    soup = bs(html.content)
    
    for i in soup.find_all("a", class_ = "core"):
        try:
            t['old_price'].append(i.find("div", class_ = "old").text)
        except:
            t['old_price'].append(np.nan)
        try:
             t['new_price'].append(i.find("div", class_ = "prc").text)
        except:
                t['new_price'].append(np.na)
        try:
              t['all_items_name'].append(i.find("h3", class_ = "name").text)
        except:
                t['all_items_name'].append(np.nan)
        try:
              t['discount'].append(i.find("div", class_ = "bdg _dsct _sm").text)
        except:
                t['discount'].append(np.nan) 
        try:
             t['Rating'].append(i.find("div", class_ = "stars _s").text)
        except:
                t['Rating'].append(np.nan)
        try:
             t["num_of_people_rated"].append(i.find("div", class_ = "rev").text.split('(')[1].strip(')'))
        except:
            t["num_of_people_rated"].append(np.nan)
    page+=1
```

```python
df = pd.DataFrame(t)
df
```

# Waterstones website

```python
from selenium import webdriver # for controlling web browser

from selenium.webdriver.chrome.service import Service as ChromeService

# part of the Selenium Library for configuring Chrome browser settings.
import requests
from bs4 import BeautifulSoup

y = {"author_of_book": [],
     "title_of_book" :[],
     "cover_type" :[],
    "new_price": [],
      "old_price" : [],
    "stock_availability": []}

page = 1
while page <=50:
    url = 'https://www.waterstones.com/books/bestsellers/sort/bestselling/page/{}'.format(page)
    path = 'C:/Users/KENNETH/Documents/LASOP DATA SCIENCE/SOFTWARE/chromedriver-win64/chromedriver.exe'


    # set up the Chrome webdriver
    chrome_service = ChromeService(executable_path=path)
    driver = webdriver.Chrome(service=chrome_service)

    # Open the webpage
    driver.get(url)

    # wait for the page to load(you might need to adjust the waiting time)
    #driver.implicitly_wait(10)

    # # Get the page source after Javascript execution
    page_source = driver.page_source

    soup =BeautifulSoup(page_source, 'html.parser')

    import pandas as pd

    for i in soup.find_all('div', class_='info-wrap'):
        try:
            y["author_of_book"].append(i.find('span', class_='author').text.strip())
        except:
             y["author_of_book"].append(np.nan)

        try:
            y["title_of_book"].append(i.find('div', class_='title-wrap').text.strip())
        except:
              y["title_of_book"].append(np.nan)

        try:
            y["cover_type"].append(i.find('div', class_='book-price').find_all("span")[-1].text.strip())
        except:
            y["cover_type"].append(np.nan)

        try:
            y["new_price"].append(i.find('span', class_="price").text)
        except:
            y["new_price"].append(np.nan)

        try: 
            y["old_price"].append(i.find('span', class_='price-rrp').text)
        except:
            y["old_price"].append(np.nan)

        try:
             y["stock_availability"].append(i.find('span', class_="format").text)
        except:
             y["stock_availability"].append(np.nan)


    page +=1
    
df= pd.DataFrame(y)
df.to_csv("book.csv", index=False)
```
