```python
t = {"num_of_people_rated":[], 'Rating':[], 'discount': [], 'new_price':[], 'old_price':[], 'all_items_name':[]}
```

```python
try:
    for people in num_of_people_rated:
        t['num_of_people_rated'].append(people.text.split('(')[1].strip(')'))

except: 
     t['num_of_people_rated'].append(np.nan)
    
try:
    for i in rating:
        t['Rating'].append(i.text)

except: 
     t['Rating'].append(np.nan)

try:
    for i in discount:
        t['discount'].append(i.text)

except: 
     t['discount'].append(np.nan)
        
try:
    for i in new_price:
        t["new_price"].append(i.text)
    
except:
     t["new_price"].append(np.nan)
        
try:
    for old in old_price:
        t["old_price"].append(old.text)

except:
    t["old_price"].append(np.nan)
    
try:
    for i in all_items_name:
        t['all_items_name'].append(i.text)

except:
     t['all_items_name'].append(np.nan)
```

```python
pd.DataFrame({key:pd.Series(value) for key, value in t.items()})
```

# To find other pages
when you copied the url, change the maximum number to {i}

```python
h = 1
t = {"num_of_people_rated":[], 
     'Rating':[], 
     'discount': [], 
     'new_price':[], 
     'old_price':[], 
     'all_items_name':[]}

while h <= 50:
    url = "https://www.jumia.com.ng/catalog/?q=shoes&page={}#catalog-listing".format(h)
    html = requests.get(url)
    soup = bs(html.content)
    
    for i in soup.find_all("a", class_ = "core"):
        try:
            t['old_price'].append(i.find("div", class_ = "old").text)
        except:
            t['old_price'].append(np.nan)
        try:
             t['new_price'].append(i.find("div", class_ = "prc").text)
        except:
                t['new_price'].append(np.na)
        try:
              t['all_items_name'].append(i.find("h3", class_ = "name").text)
        except:
                t['all_items_name'].append(np.nan)
        try:
              t['discount'].append(i.find("div", class_ = "bdg _dsct _sm").text)
        except:
                t['discount'].append(np.nan) 
        try:
             t['Rating'].append(i.find("div", class_ = "stars _s").text)
        except:
                t['Rating'].append(np.nan)
        try:
             t["num_of_people_rated"].append(i.find("div", class_ = "rev").text.split('(')[1].strip(')'))
        except:
            t["num_of_people_rated"].append(np.nan)
    h+=1
```

```python
yy=pd.DataFrame(t)
yy.to_csv("shoe.csv", index=False)
```

```python
# Open the webpage
driver.get(url)
```
